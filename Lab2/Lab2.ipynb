## 1) Write the R probability functions for the probability density and cumulative distribution functions, following the R naming convention

```r
k <- c(1,2,3,4,5)

dunif.2 <- function(n) {
    duni2 <- ifelse(n>0 & n<6, 
            n/15, 0)
    return (duni2)
    }
t <- -5:15
z <- dunif.2(t)
ggplot() + geom_point(aes(x=t, y=z))
```
